# Decision Tree Regressor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes
```

## Importing Data

```python
data = load_diabetes()
```

```python
data.feature_names
```

```python
x = data.data
y = data.target
```

```python
len(x)
```

## Training Decision Tree

```python
regressor = DecisionTreeRegressor(max_depth=3)
regressor.fit(x, y)
```

## Visualizing Decision Tree

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 7))
plot_tree(regressor, filled=True, feature_names=data.feature_names)
plt.show()
```

## Visualizing Dataset and Decision Boundaries

```python
plt.plot(x[:, 8], y,  "ro")
plt.xlabel("s5")
plt.ylabel("y")
plt.plot([-0.1, -0.004], [109.986, 109.986], "k-", linewidth=2)
plt.plot([-0.004, 0.10], [193.152, 193.152], "k-", linewidth=2)
plt.plot([-0.004, -0.004], [0, 350], "k--", linewidth=2)
plt.show()
```
