# Decision Tree Classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as pltimage
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

## Importing Data

```python
data = load_iris()
```

```python
data.feature_names
```

```python
data.target_names
```

```python
x = data.data
y = data.target
```

## Training Decision Tree

```python
classifier = DecisionTreeClassifier(max_depth=3)
classifier.fit(x, y)
```

## Visualizing Decision Tree

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 7))
plot_tree(classifier, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.show()
```

## Visualizing Dataset and Decision Boundaries

```python
plt.figure(figsize=(8, 4))
plt.plot(x[:, 2][y==0], x[:, 3][y==0], "yo", label="Iris-Setosa")
plt.plot(x[:, 2][y==1], x[:, 3][y==1], "bs", label="Iris-Versicolor")
plt.plot(x[:, 2][y==2], x[:, 3][y==2], "g^", label="Iris-Virginica")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=2)
plt.plot([2.45, 7.5], [1.75, 1.75], "k--", linewidth=2)
plt.plot([4.95, 4.95], [0, 1.75], "k:", linewidth=2)
plt.plot([4.85, 4.85], [1.75, 3], "k:", linewidth=2)
plt.text(1.40, 1.0, "(Depth=0)", fontsize=13)
plt.text(3.2, 1.80, "(Depth=1)", fontsize=12)
plt.text(4.05, 0.5, "(Depth=2)", fontsize=11)
plt.show()
```
